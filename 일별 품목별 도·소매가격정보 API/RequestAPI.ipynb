```python
import urllib.request
import json
import pandas as pd

url = 'http://www.kamis.or.kr/service/price/xml.do?action=periodProductList'

# key and id are private
parameters = '&p_cert_key=' + '' \
             + '&p_cert_id=' + '' \
             + '&p_returntype=' + 'json' \
             + '&p_startday=' + '2020-01-01' \
             + '&p_endday=' + '2020-12-31' \
             + '&p_productclscode=' + '01' \
             + '&p_itemcategorycode=' + '200' \
             + '&p_itemcode=' + '245' \
             + '&p_kindcode=' + '00' \
             + '&p_productrankcode=' + '05' \
             + '&p_countrycode=' + '1101' \
             + '&p_convert_kg_yn=' + 'Y'
url += parameters

response = urllib.request.urlopen(url)
json_str = response.read().decode("utf-8")
json_object = json.loads(json_str)
json_object
```

```python
json_object['data']['item']

import pandas as pd
df_data = pd.DataFrame(json_object['data']['item'])
df_data
df_data.to_csv("example.csv", encoding="utf-8")
```
